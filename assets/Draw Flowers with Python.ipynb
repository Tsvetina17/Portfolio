```python
# Import statements
from turtle import *
import colorsys

import numpy as np
import matplotlib.pyplot as plt
```

## Function 1
Note that the below function can be run multiple times. I experimented with different colors and layers. For example, the following two flowers are result of combining *skyblue* and *blue* colors (left) and layering *blue* three times (right).



```python
# First function for drawing flowers
def flower1():
    '''
    Draws simple flowers by taking user input for background and color of the flower.
    Can be rerun multiple times with different colors for the flower.
    '''

    # Ask for user input
    color_background = str(input('What color do you choose for the background?')) # white
    color_flower = str(input('What color do you choose for the flower?')) # skyblue
    
    # Loop that draws the flower 
    for i in range(100):
        bgcolor(color_background)
        color(color_flower)
        speed(700)
        circle(190 - i, 90)
        left(90)
        circle(190 - i, 90)
        left(190)

flower1()
hideturtle()
mainloop()
```

## Function 2
This function uses a bigger pen size and draws a flower with a yellow circle at the center.



```python
# Second function for drawing flowers
def flower2():
    '''
    Draws simple flowers by taking user input for background and color of the flower.
    By default the function draws a yellow circle in the center of the flower.
    '''

    # Ask for user input
    color_background = str(input('What color do you choose for the background?')) # white
    color_flower = str(input('What color do you choose for the flower?')) # orange

    # Loop that draws the flower 
    for i in range(100):
        goto(0,0)
        pensize(5)
        bgcolor(color_background)
        color(color_flower)
        speed(700)
        circle(230 - i, 90)
        left(90)
        circle(190 - i, 90)
        left(190)

    # Draw a yellow circle in the center of the flower
    # Initialize the turtle in the center of the flower
    goto(-90,0)
    color('yellow')
    begin_fill()
    circle(90,360)
    end_fill() 

flower2()
hideturtle()
mainloop()
```

## Function 3
The third function allows the user to experiment with different values not only for the color of the flower and background but also pen size, repetitions and extent of the circles that determine the shape of the flowers.


```python
# Third function for drawing flowers
# Ask for user input
extent_circle1 = int(input('Choose extent parameter #1')) # 50 / 30
extent_circle2 = int(input('Choose extent parameter #2')) # 70 / 50
repetitions = int(input('Choose the number of repetitions')) # 100 / 50
pen_size = int(input('Select pen size')) # 3 / 1

def flower3(extent1, extent2, reps, pen):
    '''
    Takes user input for the color of the flower, color of the background,
    pen size, repetitions and extent of the circles that determine the shape of the flowers.
    '''

    # Ask for user input
    color_background = str(input('What color do you choose for the background?')) # white
    color_flower = str(input('What color do you choose for the flower?')) # green / purple
    for i in range(reps):
        goto(0,0)
        pensize(pen)
        bgcolor(color_background)
        color(color_flower)
        speed(700)
        circle(230 - i, extent1)
        left(90)
        circle(190 - i, extent2)
        left(190)

flower3(extent_circle1, extent_circle1, repetitions, pen_size)
hideturtle()
mainloop()
```

## Function 4
Colorsys allows this function to draw very colorful flowers. I choose a hue of red to start with and experiment with different values for the parameters until I successfully print out a rainbow flower.



```python
def flower4():
    '''
    Takes user input for the background and flower. 
    It uses the entered color hue for the flower as a starting point and alternates between different hues.
    '''

    # Ask for user input
    color_background = str(input('What color do you choose for the background?')) # white
    h = float(input('What starting hue do you choose?')) # 20

    bgcolor(color_background)
    speed(700)
    
    for i in range(140):
        c = colorsys.hsv_to_rgb(h, 1, 1)
        h += 0.006
        fillcolor(c)
        begin_fill()
        circle(150 - i, 90)
        lt(20)
        lt(90)
        circle(150 - i, 90)
        rt(80)
        end_fill()

flower4()
hideturtle()
mainloop()
```

## Function 5
In the final experiment with turtle, I make another rainbow flower but increase the rotations. The result is below.



```python
def flower5():
    '''
    Takes user input for the background and flower. 
    It uses the entered color hue for the flower as a starting point and alternates between different hues.
    '''

    # Ask for user input
    color_background = str(input('What color do you choose for the background?')) # white
    h = float(input('What starting hue do you choose?')) # 10

    bgcolor(color_background)
    speed(700)
    
    for i in range(200):
        c = colorsys.hsv_to_rgb(h, 1, 1)
        h += 0.004
        fillcolor(c)
        begin_fill()
        circle(150 - i, 90)
        lt(20)
        lt(90)
        circle(150 - i, 90)
        rt(80)
        end_fill()

flower5()
hideturtle()
mainloop()
```

## Function 6
I use rose curves to draw the flower petals. The function takes user input for the number of petals, the angle used to draw the petals and, just like the previous functions, for the color of the background and the petals. You can see the output of the function below.

```python
def rose_curve(theta, n_petals):
    r = np.sin(n_petals * theta)
    return r

def flower6():
    '''
    Uses rose curves to draw the flower petals.
    Takes user input for the number of petals, the angle used to draw the petals,
    color of the background and color of the petals. 
    '''

    # Generate theta values (angle in radians)
    theta = np.linspace(0, 2*np.pi, 1000)

    # Calculate the rose curve
    r = rose_curve(theta, n)

    # Convert polar coordinates to Cartesian coordinates
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    return x, y

# Create the plot
plt.figure(figsize = (5,5))

color_flower = str(input('What color do you choose for the flower?')) # turquoise
n = int(input('What n (number of petals) do you choose?')) # 40
theta = int(input('What theta (angle in radians) do you choose?')) # 30

x, y = flower6()
plt.plot(x, y, color = color_flower)

plt.axis('off')
plt.show()
```

## Function 7
Function 7 is very similar to function 6 but it is rewritten so that it can be easily run miltuple times and the output can be used as different layers of the same flower.

```python
def rose_curve(theta, n_petals):
    r = np.sin(n_petals * theta)
    return r

def flower7():
    '''
    Uses rose curves to draw the flower petals.
    Takes user input for the number of petals, the angle used to draw the petals,
    color of the background and color of the petals. 
    It can be rerun multiple times to produce different layers of the same flower.
    '''

    color_flower = str(input('What color do you choose for the flower?')) # purple
    n = int(input('What n (number of petals) do you choose?')) # 31
    theta = int(input('What theta (angle in radians) do you choose?')) # 30

    # Generate theta values (angle in radians)
    theta = np.linspace(0, 2*np.pi, 1000)

    # Calculate the rose curve
    r = rose_curve(theta, n)

    # Convert polar coordinates to Cartesian coordinates
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    plt.plot(x, y, color = color_flower)

    return x, y, color_flower

# Create the plot
plt.figure(figsize = (5,5))

flower7() # purple, 31, 30
flower7() # pink, 21, 30
flower7() # magenta, 17, 30

plt.axis('off')
plt.show()
```
